```python
import numpy as np
import matplotlib.pyplot as plt
for pr in np.arange(start=0.2, stop=.6, step=.1):
    x = np.arange(start=1, stop=20)
    y = (1 - pr)**x
    plt.plot(x,y, label = f"{pr.round(2)}")
plt.plot(np.ones_like(x)*0.05)
plt.legend()
plt.show()
```

```python
pr = 0.5
x = np.arange(start=0, stop=20)
y = (1 - pr)**x
# y = np.sin(x)
print(y)
xvals = np.linspace(0, 20, 4)

yinterp = np.interp(xvals, x, y)

import matplotlib.pyplot as plt

plt.plot(x, y, 'o')
plt.plot(xvals, yinterp, '-x')
plt.show()
```
