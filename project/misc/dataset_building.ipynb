```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets 

def plot_dataset(a,b,c):
    """a is the x axis, b is the y-axis, c is the class label"""
    # Plotting the complete dataset
    plt.figure(figsize=(4, 4))
    plt.scatter(a, b, c=c, cmap=plt.cm.RdYlBu)
    plt.show()

def concat_datasets(*datasets):
    """
    concatenate datasets with shapes
    X : (N, d)
    Y : (N, )

    returns X : (N, #d) Y : (N, #)
    """
    x,y = list(zip(*datasets))
    X = np.hstack(x)
    Y = np.vstack(y).T

    return X, Y

n_samples = 500
noise = 0.05
X, y = dataset = concat_datasets(
    #datasets.make_circles(n_samples, noise=noise, random_state=170, shuffle=True),
    datasets.make_moons(n_samples, noise=noise, random_state=0, shuffle=True),
    datasets.make_moons(n_samples, noise=noise, random_state=0, shuffle=True)
)

# Load the dataset

task_1 = X[:, 0], X[:, 1], y[:,0]
task_2 = X[:, 2], X[:, 3], y[:,1]
mixed_samples_11 = X[:, 1], X[:, 2], y[:,1]
mixed_samples_12 = X[:, 1], X[:, 2], y[:,0]
mixed_samples_21 = X[:, 0], X[:, 3], y[:,1]
mixed_samples_22 = X[:, 0], X[:, 3], y[:,0]
wrong_labels_1 = X[:, 0], X[:, 1], y[:,1]
wrong_labels_2 = X[:, 2], X[:, 3], y[:,0]

plot_dataset(*task_1)
plot_dataset(*task_2)

plot_dataset(*mixed_samples_11)
plot_dataset(*mixed_samples_12)
plot_dataset(*mixed_samples_21)
plot_dataset(*mixed_samples_22)

plot_dataset(*wrong_labels_1)
plot_dataset(*wrong_labels_2)
```

```python
from sklearn import datasets
import numpy as np

# Define a function to concatenate multiple datasets
def concatenate_datasets(*datasets):

    # Initialize empty arrays for the combined X and y
    combined_X = np.empty((0, datasets[0][0].shape[1]))
    combined_y = np.empty(0, dtype=int)

    # Loop through each dataset and concatenate
    for X, y in datasets:
        combined_X = np.vstack((combined_X, X))
        combined_y = np.concatenate((combined_y, y))

    return combined_X, combined_y

# Create the first dataset
X1, y1 = datasets.make_circles(n_samples=500, noise=0.05, random_state=170, shuffle=True)

# Create additional datasets (as examples)
X2, y2 = datasets.make_circles(n_samples=500, noise=0.05, random_state=42, shuffle=True)
X3, y3 = datasets.make_circles(n_samples=500, noise=0.05, random_state=123, shuffle=True)

# Concatenate the datasets
combined_X, combined_y = concatenate_datasets((X1, y1), (X2, y2), (X3, y3))

datasets = [
    datasets.make_circles(n_samples=10, noise=0.05, random_state=42, shuffle=True),
    datasets.make_circles(n_samples=10, noise=0.05, random_state=123, shuffle=True)
]
x,y = list(zip(*datasets))

X = np.hstack(x)
Y = np.vstack(y).T


assert X.shape == (10,4), X.shape
assert Y.shape == (10,2), Y.shape

assert X[0,0] == x[0][0,0]
assert X[0,2] == x[1][0,0]
assert X[0,3] == x[1][0,1]

assert Y[0,0] == y[0][0]
assert Y[9,0] == y[0][9]
assert Y[9,1] == y[1][9]
```
